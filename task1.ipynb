```python
# # define parameters for the system
# evolution operator for the linear system
alpha = 0.1
A = np.array([
    [alpha, alpha],
    [-1/4, 0]
])
# system domain
x_min = -1
x_max = 1
n_x = 100
y_min = -1
y_max = 1
n_y = 100
list_par = [
    [x_min, x_max, n_x],
    [y_min, y_max, n_y]
]
# # define the system
system = Task1(par=A, list_par_x=list_par)
X = system.X
DX = system.fun(0, X) # t is arbitrary here, since this system is time-independent
# # plot the phase portrait
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
fig, ax = plot_phase_portrait(fig, ax, X, DX)
# # plot a trajectory
# define parameters for solving the system
init_state = np.array([0, 1e-2])
t_eval = np.linspace(0, 100, 1000)
# solve the system
trajectory = system.solve_system(
    fun=system.fun,
    init_state=init_state,
    t_eval=t_eval
)
# plot trajectory
ax.plot(trajectory[0], trajectory[1])
```
